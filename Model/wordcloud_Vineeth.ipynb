```python
import pandas as pd
import os
from tqdm import tqdm

```

```python
current_path = os.getcwd()+'/'
dataset_path =  current_path+'Collected Datasets/'
```

```python
df_train = pd.read_csv(dataset_path+'train.csv')
df_dev = pd.read_csv(dataset_path+'devel.csv')
df_test = pd.read_csv(dataset_path+'test.csv')
```

```python
#this function enables to pick out on the users interested columns  
intestested_columns = ['message', 'emotion']
df_train = df_train[intestested_columns]
df_dev = df_dev[intestested_columns]
df_test = df_test[intestested_columns]
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#this function picks out the 100 words from each dataset, it sets the top frequency 
#and picks out the words from the datasets
anger_sentences = ' '.join(df_train[df_train['emotion']=='anger']['message'].values.tolist())
sad_sentences = ' '.join(df_train[df_train['emotion']=='sadness']['message'].values.tolist())
joy_sentences = ' '.join(df_train[df_train['emotion']=='joy']['message'].values.tolist())
fear_sentences = ' '.join(df_train[df_train['emotion']=='fear']['message'].values.tolist())
normal_sentences = ' '.join(df_train[df_train['emotion']=='normal']['message'].values.tolist())
```

```python
#this function displays the words based on the emotion anger
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(anger_sentences)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
